## Cleaning steps

```r
# Import our Datasets
file_path1 <- "/content/Jon Stewart on Trump’s Inauguration and Elon Musk_s Nazi Salute _ The Daily Show - YouTube.xlsx"
file_path2 <- "/content/Elon Musk_ War, AI, Aliens, Politics, Physics, Video Games, and Humanity _ Lex Fridman Podcast _400 .xlsx"

# Import the Excel file
df_Show <- read_excel(file_path1)
df_Podcast <- read_excel(file_path2)

# Display the first rows of the imported file
head(df_Show)
head(df_Podcast)
```

```r
# Add a source column to the datasets to prevent loss of information
df_Podcast$source <- "Podcast"
df_Show$source <- "Show"
```

```r
# Combine both datasets (cleaning will be more efficent this way)
df_combined <- rbind(df_Podcast, df_Show)
```

```r
# Check NA values
sum(is.na(df_combined)) #it seems that comments with no replies or likes have NA values.
```

```r
# Replace NA by 0 in specific columns
df_combined$number_of_likes[is.na(df_combined$number_of_likes)] <- 0
df_combined$number_of_responses[is.na(df_combined$number_of_responses)] <- 0
```

```r
# Remove rows containing non-Latin characters
df_combined <- df_combined %>%
  filter(grepl("[A-Za-z]", comment))

# Check the result
head(df_combined)
```

```r
# Replace NA values with an empty string
df_combined$comment[is.na(df_combined$comment)] <- ""
```

# Clustering steps

```r
# Create the corpus from the "comment" column
text_corpus <- Corpus(VectorSource(df_combined$comment))

# Text cleaning
text_corpus <- tm_map(text_corpus, content_transformer(function(x) iconv(x, to = "ASCII", sub = "byte"))) # Convert to ASCII
text_corpus <- tm_map(text_corpus, content_transformer(tolower))  # Lower case
text_corpus <- tm_map(text_corpus, removePunctuation)             # Removing punctuation
text_corpus <- tm_map(text_corpus, removeNumbers)                 # Removing numbers
text_corpus <- tm_map(text_corpus, removeWords, stopwords("en"))  # Removing stopwords
text_corpus <- tm_map(text_corpus, stripWhitespace)               # Removing unnecessary blank spaces
```

```r
# Exclude certain words
excluded_words<- c("elon", "lex", "really", "trump", "interview", "one", "just", "thank", "jon", "musk", "can", "time", "will", "know","que")
```

```r
# Words must be considered as separated entities

df_tokens<- df_combined %>%
  unnest_tokens(word, comment) %>%
filter(!word %in% excluded_words)
```

```r
# Apply text cleaning and word exclusion in the corpus
text_corpus <- tm_map(text_corpus, content_transformer(function(x) {
  removeWords(x, excluded_words)
}))

# Remove extra spaces after word removal
text_corpus <- tm_map(text_corpus, stripWhitespace)
```

```r
# Create the Document-Term Matrix
dtm <- TermDocumentMatrix(text_corpus)

# Convert to a matrix
m <- as.matrix(dtm)

# Display the first 5 terms for the first 5 documents
m[1:5, 1:5]
```

```r
v <- sort(rowSums(m),decreasing=TRUE)
```

```r
# Assume that `v` contains the word frequencies extracted from the Document-Term Matrix (DTM)
d <- data.frame(word = names(v), freq = v)

# Display the first 5 rows of the dataframe
head(d, 5)
```

```r
d <- d[!d$word %in% excluded_words, ]

# Generate the wordcloud with the new filtered list
wordcloud(d$word, d$freq, min.freq = 15, random.order=FALSE, colors = brewer.pal(8, "Dark2"))
```

```r
# Generate a barplot
barplot(d[1:20,]$freq, las = 3,
names.arg = d[1:20,]$word,col ="lightblue", main ="Most frequent words",
ylab = "Word frequencies")
```

```r
findFreqTerms(dtm, lowfreq=0.50)[1:10]
```

```r
#Remove sparse terms (these terms are not present enough to be meaningful, but they could slow down the next steps)
dtm.sparse<-removeSparseTerms(dtm, sparse=0.97)
```

```r
#Convert the term-document matrix to a data frame
dtm.sparse <- as.data.frame(as.matrix(dtm.sparse))

#Scaling
dtm.sparse.scale<- scale(dtm.sparse)
```

```r
dtm.sparse.dist<- dist(dtm.sparse.scale, method = "euclidean")
```

```r
length(dtm.sparse.dist)
```

```r
# Cluster the data
dtm.sparse.fit<-hclust(dtm.sparse.dist, method="ward.D2")
```

```r
#Clustering plot
groups <- cutree(dtm.sparse.fit, k=2)

plot(dtm.sparse.fit, main="Cluster-ElonMusk")

rect.hclust(dtm.sparse.fit, k=2, border="red")
```

# Sentiment Analysis

```r
install.packages("syuzhet")
library(syuzhet)

# Initialize an empty vector to store comments
plain.text <- vector(mode = "character", length = nrow(df_combined))

# Loop to extract each comment
for(i in 1:nrow(df_combined)){
  plain.text[i] <- df_combined$comment[i]
}

# Apply the sentiment analysis
sentence_sentiment <- get_sentiment(get_sentences(plain.text))

# Display the results
sentence_sentiment
```

```r
# Calculate the average sentiment scores
average_sentiment <- mean(sentence_sentiment)

# Display the result
average_sentiment
```

```r
# Calculate the standard deviation of sentiment scores
sd_sentiment <- sd(sentence_sentiment)

# Display the result
sd_sentiment
```

```r
#Sentiment terms

# Extract sentences from the text
sentences <- get_sentences(plain.text)

# Apply NRC analysis to obtain emotions and sentiment
sentiment_terms <- get_nrc_sentiment(sentences)

# Display the results
head(sentiment_terms)
```

```r
# Display all the table
sentiment_terms
```

```r
# Define positive and negative emotions
positive_emotions <- c("joy", "anticipation", "trust", "surprise", "positive")
negative_emotions <- c("anger", "disgust", "fear", "sadness", "negative")

# Calculate the sum of values for each sentiment
sentiment_sums <- colSums(sentiment_terms)

# Sort the values in descending order
sentiment_sums <- sort(sentiment_sums, decreasing = TRUE)

# Define colors based on the sorted emotions
colors <- ifelse(names(sentiment_sums) %in% positive_emotions, "green", "red")

# Create a barplot
barplot(sentiment_sums,
        col = colors,
        main = "Sentiment distribution ",
        xlab = "Sentiments",
        ylab = "Sum of apparitions",
        las = 2)
```
